```python
import pandas as pd
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/border entry.csv')
print("Shape after reading CSV:", df.shape)

# Removed filtering and date creation to prevent empty DataFrame with current data
# df = df[df['Month'] != 'Total']
# print("Shape after filtering 'Total' months:", df.shape)
# df['Date'] = pd.to_datetime(df['Year'].astype(str) + '-' + df['Month'].astype(str) + '-01', errors='coerce')
# df.dropna(subset=['Date'], inplace=True)
# print("Shape after dropping rows with invalid dates:", df.shape)
# df = df.sort_values('Date')

# Display the first few rows to show the data content
display(df.head())
```

```python
# Feature Engineering
# Ensure 'Year' is integer
df['Year'] = df['Year'].astype(int)

# Encode 'Month' column
# Note: This will encode 'Total' as well, which might affect model interpretation
df['Month_Encoded'] = LabelEncoder().fit_transform(df['Month'])
```

```python
# Define features and target
features = ['Border', 'Entry Type', 'Year', 'Month_Encoded']
target = 'Count'

X = df[features]
y = df[target]

print("Features and target defined.")
print("Features (X) shape:", X.shape)
print("Target (y) shape:", y.shape)
```

```python
# Plot predictions
plt.figure(figsize=(12, 6))
plt.plot(y.values, label='Actual')
plt.plot(y, label='Predicted')
plt.legend()
plt.title('Border Crossing Prediction')
plt.xlabel('Time Index')
plt.ylabel('Number of Crossings')
plt.show()
```
